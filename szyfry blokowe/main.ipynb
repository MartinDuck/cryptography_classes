```python
from helpers import generate_file_of_size, prepare_encryption, encrypt_data, decrypt_data, flip_random_bits_in_middle
import os   
```

```python
file_sizes = [100, 300 ,600]
modes = ["ECB", "CBC", "CFB", "OFB", "CTR"]
key = os.urandom(16)  # AES-128 key
```

```python
for size in file_sizes:
    generate_file_of_size(f"./inputs/input_{size}.txt", 1024 * 1024 * size) 
```

```python
for mode_name in modes:

    print(f"\n ------Testing mode: {mode_name}--------")
    for size in file_sizes:
            
            
            file_to_encrypt, mode, iv_or_nonce = prepare_encryption(f"./inputs/input_{size}.txt", mode_name)
            encryption_time, encrypted_data = encrypt_data(key=key, data_to_encrypt=file_to_encrypt, iv_or_nonce=iv_or_nonce, mode=mode)
            print(f"Encryption time for {size} MB: {encryption_time:.4f} seconds")

            decryption_time, decrypted_data = decrypt_data(key=key, encrypted_file=encrypted_data, mode=mode_name)
            print(f"Decryption time for {size} MB: {decryption_time:.4f} seconds")
```

```python
test_image = "inputs/panda_1024.bmp"

for mode_name in modes:
    file_to_encrypt, mode, iv_or_nonce = prepare_encryption(test_image, mode_name)
    encryption_time, encrypted_data = encrypt_data(key=key, data_to_encrypt=file_to_encrypt, iv_or_nonce=iv_or_nonce, mode=mode)

    encrypted_data = flip_random_bits_in_middle(encrypted_data, num_bits=len(file_to_encrypt) // 10)  #Flip 10% of the bits in the encrypted data

    decryption_time, decrypted_data = decrypt_data(key=key, encrypted_file=encrypted_data, mode=mode_name, output_filepath=f"./outputs/decrypted_{mode_name}_panda.bmp")
```
