```python
import numpy as np
from PIL import Image

def text_to_binary(text):
    """Konwertuje tekst na ciąg zer i jedynek (8 bitów na znak)."""
    return ''.join(format(ord(c), '08b') for c in text)

def binary_to_text(binary_str):
    """Konwertuje ciąg zer i jedynek z powrotem na tekst ASCII."""
    chars = [chr(int(binary_str[i:i+8], 2)) for i in range(0, len(binary_str), 8)]
    return ''.join(chars)

def hide_message(image_path, secret_message, output_path):
    """Ukrywa wiadomość w najmniej znaczących bitach (LSB) obrazu."""
    img = Image.open(image_path)
    # Zapewniamy, że obraz to RGB
    img = img.convert('RGB')
    pixels = np.array(img)
    
    binary_message = text_to_binary(secret_message) + '0000000000000000'
    message_length = len(binary_message)
    
    flat_pixels = pixels.flatten()
    
    if message_length > len(flat_pixels):
        raise ValueError("Obraz jest zbyt mały, aby pomieścić tę wiadomość!")
        
    # Zmiana LSB
    for i in range(message_length):
        flat_pixels[i] = (flat_pixels[i] & 254) | int(binary_message[i])
        
    new_pixels = flat_pixels.reshape(pixels.shape)
    new_img = Image.fromarray(new_pixels.astype(np.uint8), 'RGB')
    new_img.save(output_path)
    print(f"Pomyślnie ukryto wiadomość. Zapisano do: {output_path}")
        

    new_pixels = flat_pixels.reshape(pixels.shape)
    new_img = Image.fromarray(new_pixels.astype(np.uint8), 'RGB')
    new_img.save(output_path)
    print(f"Pomyślnie ukryto wiadomość. Zapisano do: {output_path}")

def extract_message(image_path):
    """Wydobywa wiadomość z bitów LSB obrazu."""
    img = Image.open(image_path)
    img = img.convert('RGB')
    flat_pixels = np.array(img).flatten()
    
    extracted_bits = ""
    
    for pixel_value in flat_pixels:
       
        extracted_bits += str(pixel_value & 1)
        
        if extracted_bits.endswith('0000000000000000'):
            return binary_to_text(extracted_bits[:-16])
            
    return "Nie znaleziono wiadomości lub niepoprawny format."


hide_message("test.png", "Tajna wiadomosc testowa!", "obraz_z_wiadomoscia.png")
odzyskany_tekst = extract_message("obraz_z_wiadomoscia.png")
print(f"Odczytano: {odzyskany_tekst}")
```
